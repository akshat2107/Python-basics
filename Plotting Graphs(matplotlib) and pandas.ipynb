## Movie Data Visualisation challenge

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/coding-blocks-archives/machine-learning-online-2018/master/2.%20Working%20with%20Libraries/movie_metadata.csv')
```

```python
df.head(n=10)
```

```python
df.columns
```

```python
titles = df.get("movie_title")
print(titles)
```

```python
print(type(titles))
```

```python
titles = list(titles)
print(titles)
```

```python
print(titles[:5])
```

```python
freq_title = {}
for title in titles:
    length = len(title)
    if freq_title.get(length) is None:
        freq_title[length] = 1
    else:
        freq_title[length]+=1
```

```python
print(freq_title)
```

```python
x = np.array(list(freq_title.keys()))
y = np.array(list(freq_title.values()))
plt.scatter(x,y)
plt.xlabel('no. of movies')
plt.ylabel('freq_count')
plt.title("movie data visualisation")
plt.show()
```

### Plotting a normal distribution(1-d)

```python
sigma = 5
u = 40
a = np.random.randn(50)
a = sigma*a + u #standard normal distribution
plt.figure(figsize=(3,3))
plt.hist(a,10)
plt.show()

# so basically this is a normal distribution with mean = u and standard deviation = sigma
```

```python
a = np.round(a)
z = np.unique(a,return_counts=True)
print(z)
```

```python
x = a
y = np.zeros(x.shape)
plt.scatter(x,y)
plt.show()
```

## Multivariate Normal Distribution

```python
# bivariate case
mean = np.array([0.0,0.0])
cov = np.array([[1,0.0],[0.0,1]])

dist = np.random.multivariate_normal(mean,cov,500)
plt.scatter(dist[:,0],dist[:,1])

 
mean1 = np.array([5,6])
cov1 = np.array([[1,0.8],[0.8,1]])

dist = np.random.multivariate_normal(mean,cov,500)
dist1 = np.random.multivariate_normal(mean1,cov1,500)
plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist1[:,0],dist1[:,1],c='yellow')
plt.show()
```

```python
# so here we can see that mean is 0,0 and how the cocariance is affecting
```
